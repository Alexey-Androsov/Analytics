```python
# Группировка со сбросом индекса:
b = df.groupby(['dt', 'apt', 'empl', 'pay_type'])['kol'].agg(sum).reset_index()
b
```

```python
# Сводная таблица (Pivot Table) без предварительной агрегации:
c = df.pivot_table(values='roz', index=['dt', 'apt', 'empl'], columns=['pay_type'], aggfunc=sum)
c
```

```python
c = c.reset_index()
```

```python
# Преобразование из сводной таблицы (Pivot Table) в обычную таблицу:
pd.melt(c, id_vars=['dt', 'apt', 'empl'], value_vars=[15, 18], var_name='pay_type')
```
